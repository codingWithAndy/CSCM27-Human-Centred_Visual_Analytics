```python
%matplotlib inline
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
alt.renderers.enable('notebook')
```

```python
cars = data.cars()
```

```python
cars
```

```python
alt.Chart(cars).mark_point().encode(
x = 'Horsepower',
y = 'Miles_per_Gallon',
color = 'Origin',
shape = 'Origin')
```

```python
# Quantitive Q ---- a continuous real-valued quantity.
# Ordinal    O ---- a discreate ordered quantity.
# Nominal    N ---- a discreate unordered category.
# Temporal   T ---- a time or date value.
```

```python
alt.Chart(cars).mark_circle().encode(
    alt.X('Horsepower', bin=True),
    alt.Y('Miles_per_Gallon', bin=True),
    size = 'count()',
    color = 'average(Acceleration):Q'
)
```

```python
barley = data.barley()

barley
```

```python
alt.Chart(barley).mark_bar().encode(
    x = 'variety:N',
    y = 'sum(yield):Q',
    color = 'site:N',
    order = alt.Order('site', sort='ascending')
)
```

```python
alt.Chart(barley).mark_bar().encode(
    x = 'variety:N',
    y = 'sum(yield):Q',
    color = 'site:N',
    order = alt.Order('site', sort='descending')
)
```

```python
alt.Chart(barley).mark_area().encode(
    x = 'variety:N',
    y = 'sum(yield):Q',
    color = 'site:N',
    order = alt.Order('site', sort='ascending')
)
```

```python
alt.Chart(barley).mark_area().encode(
    x = 'variety:N',
    y = 'sum(yield):Q',
    color = 'site:N',
    order = alt.Order('site', sort='descending')
)
```

```python
alt.Chart(cars).mark_point().encode(
    x = 'Acceleration:Q',
    y = 'Miles_per_Gallon:Q',
    color = 'Origin:N'
)
```

```python
# Same as above but using different syntax.
alt.Chart(cars).mark_point().encode(
    alt.X('Acceleration', type='quantitative'),
    alt.Y('Miles_per_Gallon', type='quantitative'),
    alt.Color('Origin', type='nominal')
)
```

```python
base = alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q'
).properties(
    width=150,
    height=150
)

alt.vconcat(
    base.encode(color='Cylinders:Q').properties(title='Graph for Quantative'),
    base.encode(color='Cylinders:O').properties(title='Graph for Ordinal'),
    base.encode(color='Cylinders:N').properties(title='Graph for Nominal')
)
```

```python
pop = data.population.url
```

```python
base = alt.Chart(pop).mark_bar().encode(
    alt.Y('mean(people):Q', axis=alt.Axis(title='Total Population'))
).properties(
    width=200,
    height=200
)
```

```python
alt.hconcat(
    base.encode(x='year:Q').properties(title='Year = Quantative'),
    base.encode(x='year:O').properties(title='Year = Ordinal'),
    # base.encode(x='year::N').properties(title='Year = Nominal') -> Will not work on this data.
)
```

```python
alt.Chart(cars).mark_bar().encode(
    alt.X('Horsepower', bin=True),
    alt.Y('count()')
)
```

```python
alt.Chart(cars).mark_bar().encode(
    alt.X('Horsepower', bin=True),
    alt.Y(aggregate='count', type='quantitative')
)
```

```python
driving = data.driving()
```

```python
driving
```

```python
# Vis Diagram using Scale
alt.Chart(driving).mark_line(point=True).encode(
    alt.X('miles', scale=alt.Scale(zero=False)),
    alt.Y('gas', scale=alt.Scale(zero=False)),
    order='year'
)
```

```python
# Starting point from 0
alt.Chart(driving).mark_line(point=True).encode(
    alt.X('miles', scale=alt.Scale(zero=True)),
    alt.Y('gas', scale=alt.Scale(zero=True)),
    order='year'
)
```
