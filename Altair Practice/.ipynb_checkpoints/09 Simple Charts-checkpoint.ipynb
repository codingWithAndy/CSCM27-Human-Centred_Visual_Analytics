```python
%matplotlib inline
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
alt.renderers.enable('notebook')
```

```python
source = pd.DataFrame({
    'a': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'b': np.random.randn(9)
})

alt.Chart(source).mark_bar().encode(
    x='a',
    y='b'
)
```

```python
alt.Chart(source).mark_bar().encode(
    x='b',
    y='a'
)
```

```python
x, y = np.meshgrid(range(-5, 5), range(-5, 5))
z = x ** 2 + y ** 2

source = pd.DataFrame({
    'x': x.ravel(),
    'y': y.ravel(),
    'z': z.ravel()
})

alt.Chart(source).mark_rect().encode(
    x='x:O',
    y='y:O',
    color='z:Q'
)
```

# Simple Histograms

```python
source = data.movies.url

alt.Chart(source).mark_bar().encode(
    alt.X('IMDB_Rating:Q', bin=True),
    y='count()',
)
```

```python
# Sin wave
t = np.arange(100)
source = pd.DataFrame({
    't': t,
    'f(t)': np.sin(t/7)
})

alt.Chart(source).mark_line().encode(
    x='t',
    y='f(t)'
)
```

```python
# Interactive Charts

source = data.cars()

alt.Chart(source).mark_circle(size=60).encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    tooltip=['Name', 'Origin', 'Horsepower', 'Miles_per_Gallon']
).interactive()
```

```python
# Stacked Area Chart

source = data.iowa_electricity()

alt.Chart(source).mark_area().encode(
    x='year:T',
    y='net_generation:Q',
    color="source:N"
)
```

```python
# Strip Plot

source = data.cars()

alt.Chart(source).mark_tick().encode(
    x='Horsepower:Q',
    y='Cylinders:O'
)
```
