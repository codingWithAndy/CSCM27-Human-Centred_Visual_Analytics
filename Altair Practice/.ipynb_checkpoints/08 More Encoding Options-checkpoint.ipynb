```python
%matplotlib inline
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
alt.renderers.enable('notebook')
```

```python
iris = data.iris()

iris
```

```python
alt.Chart(iris).mark_point().encode(
    x='petalWidth',
    y='petalLength',
    color='species'
)
```

```python
alt.Chart(iris).mark_point().encode(
    x='petalWidth',
    y='petalLength',
    color=alt.value('green')
)
```

```python
alt.Chart(iris).mark_point().encode(
    x='petalWidth',
    y='petalLength',
    color=alt.value('green'),
    size='count()'
)
```

```python
alt.Chart(iris).mark_point().encode(
    x='petalWidth',
    y='petalLength',
    color=alt.value('blue'),
    size=alt.value(6)
)
```
