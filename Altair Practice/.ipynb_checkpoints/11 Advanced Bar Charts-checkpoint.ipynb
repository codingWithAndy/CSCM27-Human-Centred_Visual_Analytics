```python
%matplotlib inline
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
alt.renderers.enable('notebook')
```

```python
# Horizontal Bar Chart

source = data.wheat()

alt.Chart(source).mark_bar().encode(
    x='wheat:Q',
    y='year:O'
).properties(height=700)
```

```python
# Grouped Horizontal Bar Chart

source = data.barley()

alt.Chart(source).mark_bar().encode(
    x='sum(yield)',
    y='year:O',
    color='year:N',
    row='site:N'
)
```

```python
# Horizontal Bar chart stacked

alt.Chart(source).mark_bar().encode(
    x='sum(yield)',
    y='variety',
    color='site'
)
```

```python
# Layered Bar Chart
source = data.iowa_electricity()

alt.Chart(source).mark_bar(opacity=0.7).encode(
    x='year:O',
    y=alt.Y('net_generation:Q', stack=None),
    color='source'
)
```

```python
# Normalized Horizontal Stack Bar Chart
source = data.barley()

alt.Chart(source).mark_bar().encode(
    x=alt.X('sum(yield)', stack='normalize'),
    y='variety',
    color='site'
)
```

```python
# Sorted Bar Chart

alt.Chart(source).mark_bar().encode(
    x=alt.X('sum(yield):Q'),
    y=alt.Y(
        'site:N',
        sort=alt.EncodingSortField(
            field='yield',      #
            op='sum',
            order='descending'
        )
    )
)
```

```python
# Vertical Stacked Bar Chart
alt.Chart(source).mark_bar().encode(
    x='variety',
    y='sum(yield)',
    color='site'
)
```

```python
# Horizontal Stacked bar chart with sorted segments
alt.Chart(source).mark_bar().encode(
    x='sum(yield)',
    y='variety',
    color='site',
    order=alt.Order(
        'site',
        sort = 'ascending'
    )
)
```

```python
# Horizontal Trellis Stacked Bar Chart

alt.Chart(source).mark_bar().encode(
    column='year',
    x='sum(yield)',
    y='variety',
    color='site'
).properties(width=250)
```
