```python
import altair as alt #imports Altair into Notebook.
```

```python
%matplotlib inline  
#Allows us to see Matplotlib within the notebook -> Altair is a library for Matplot
alt.renderers.enable('notebook') #This enables altair to work in Notebook.
```

```python
alt.Chart(iris).mark_point().encode(
x = 'petalLength',
y = 'petalWidth')
```

```python
alt.Chart(iris).mark_point().encode(
x = 'petalLength',
y = 'petalWidth',
color = 'species')
```

```python
alt.Chart(iris).mark_point().encode(
x = 'sepalLength',
y = 'sepalWidth',
color = 'species')
```

```python
alt.Chart(iris).mark_point().encode(
x = 'petalLength',
y = 'petalWidth',
color = 'species',
size = 'sepalLength')
```
